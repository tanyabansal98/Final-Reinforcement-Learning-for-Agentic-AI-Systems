## 1. Project Overview
This notebook acts as the **Master Orchestrator** for the Madison RL project. It executes a modular codebase designed to demonstrate five different reinforcement learning methods working in symphony to solve information gathering and synthesis tasks.

### The 5-Layer RL Architecture

| Layer | RL Method | Role |
|:---|:---|:---|
| **1. Prioritization** | Meta-Learning (MAML) | Fast adaptation to new research domains. |
| **2. Selection** | Contextual Bandits (LinUCB) | Picking the highest probability source category. |
| **3. Sequencing** | Q-Learning & SARSA | Optimizing the order of queries in a session. |
| **4. Coordination** | Multi-Agent RL (MAPPO) | Inter-agent communication to avoid redundancy. |
| **5. Synthesis** | Policy Gradient (PPO) | Weighting results for final research output. |
